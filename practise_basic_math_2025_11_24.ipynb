**Задача 1** \
Дані — перегляд (хв): \
[3, 4, 5, 6, 7]

Знайти медіану.

```python
arr = [4, 5, 3, 7, 6]

arr.sort() # sorted просто сортує дані, без зміни самого масиву
n = len(arr)
if n % 2 == 1:
  median_ = arr[n // 2]
  print(median_)
else:
    median_ = (arr[n // 2 - 1] + arr[n // 2]) / 2
    print(median_)
```

**Задача 1** \
Дані: [10, 12, 12, 15, 20]\
Знайти моду.

```python

arr = [10, 12, 12, 15, 20]

avg_arr = sum(arr) / len(arr) # Обчислюємо середнє арифметичне елементів у списку
print(avg_arr)

ans = {} # Створюємо порожній словник для зберігання частоти кожного числа

for num in arr: # Перебираємо кожен елемент у списку
  if num not in ans: # Якщо число ще не зустрічалося
    ans[num] = 1 # Додаємо його до словника з частотою 1
  else: # Якщо число вже зустрічалося
    ans[num] +=1 # Збільшуємо його частоту на 1
print(ans) # Виводимо словник з частотами чисел


max_freq = 0 # Ініціалізуємо змінну для максимальної частоти
max_value = 0 # Ініціалізуємо змінну для значення з максимальною частотою (моди)

for key, value in ans.items(): # Перебираємо пари ключ-значення (число-частота) у словнику
  if value > max_freq: # Якщо поточна частота більша за максимальну
    max_freq = value # Оновлюємо максимальну частоту
    max_value = key # Оновлюємо значення моди
print(max_value) # Виводимо моду


from collections import Counter # Імпортуємо клас Counter з модуля collections
ans = Counter(arr) # Створюємо об'єкт Counter, який автоматично обчислює частоту елементів
print(ans.most_common(1)) # Виводимо найбільш поширений елемент (моду) та його частоту. Якщо в (1) прибрати один, метод виведе всі елементи і  їх частоту
```

Задача 2 — два набори з однаковим середнім, але різним розкидом \
A: [4, 5, 6] \
B: [1, 5, 9]

Середні однакові.  
Який набір стабільніший?

```python
a_set = [4, 5, 6]
b_set = [1, 5, 9]

# Step 1. calculate avg (обчислюємо середнє значення)

avg_a_set = sum(a_set) / len(a_set)  # рахуємо середню
avg_b_set = sum(b_set) / len(b_set)   # рахуємо середню
print(f"Середнє значення A: {avg_a_set}")
print(f"Середнє значення B: {avg_b_set}")

# Step 2. x(i) - avg (обчислюємо різницю між кожним елементом та середнім)

differences_a = []                    # створюємо порожній список для різниць a_set
for x in a_set:                       # перебираємо кожен елемент у a_set
    diff_a = x - avg_a_set              # обчислюємо x(i) - середнє
    differences_a.append(diff_a)        # додаємо різницю до списку
    print(f"diff a {diff_a}")                    # можна вивести для перевірки
print("Різниці:", differences_a)      # виводимо всі різниці a_set

differences_b = []                    # створюємо порожній список для різниць b_set
for x in b_set:                       # У ТЕБЕ БУЛА ПОМИЛКА! Потрібно перебирати b_set
    diff_b = x - avg_b_set              # обчислюємо x(i) - середнє b_set
    differences_b.append(diff_b)        # додаємо різницю до списку
    print(f"diff b {diff_b}")                       # виводимо різницю кожного елемента
print("Різниці:", differences_b)      # виводимо всі різниці b_set

# Step 3. step 2 ** 2 (квадрат кожної різниці)

qua_differences_a = []                # список квадратів різниць для a_set
for diff in differences_a:            # перебираємо всі різниці і
    qua_diff = diff ** 2              # підносимо кожну різницю до квадрату
    qua_differences_a.append(qua_diff) # додаємо до нового списку
    print(f"вводимо в кв.{qua_diff}")  # виводимо квадрат різниці

qua_differences_b = []                # список квадратів різниць для b_set
for diff in differences_b:            # перебираємо всі різниці і
    qua_diff = diff ** 2              # обчислюємо квадрат різниці
    qua_differences_b.append(qua_diff) # додаємо до списку
    print(f"вводимо в кв.{qua_diff}")  # виводимо квадрат

# Step 4. sum - step 3 (сума квадратів різниць)

sum_qua_differences_a = sum(qua_differences_a)  # сума квадратів різниць a_set
sum_qua_differences_b = sum(qua_differences_b)  # сума квадратів різниць b_set
print(f"різниця {sum_qua_differences_a}")       # виводимо суму a_set
print(f"різниця {sum_qua_differences_b}")       # виводимо суму b_set

# Step 5. step 4 / n (ділимо на кількість елементів → дисперсія)

disp_a = sum_qua_differences_a / len(a_set)     # дисперсія a_set
disp_b = sum_qua_differences_b / len(b_set)     # дисперсія b_set
print(f"дисперсія a_set: {disp_a}")             # виводимо дисперсію a_set
print(f"дисперсія b_set: {disp_b}")             # виводимо дисперсію b_set
```

### Std
Дані: [4, 5, 6, 9, 10]  

Порахуй variance і std.


```python
arr = [4, 5, 6, 9, 10]
avg_arr = sum(arr) / len(arr)  # рахуємо середню
print(f"середнє {avg_arr:.2f}")

# Step 2. x(i) - avg (обчислюємо різницю між кожним елементом та середнім)

differences = []                    # створюємо порожній список для різниць arr
for x in arr:                       # перебираємо кожен елемент у arr
    diff = x - avg_arr              # обчислюємо x(i) - середнє
    differences.append(diff)        # додаємо різницю до списку
    # print(f"diff a {diff}")                    # можна вивести для перевірки
# print("Різниці:", differences)      # виводимо всі різниці arr

# Step 3. step 2 ** 2 (квадрат кожної різниці)

qua_differences = []                # список квадратів різниць для arr
for diff in differences:            # перебираємо всі різниці і
    qua_diff = diff ** 2              # підносимо кожну різницю до квадрату
    qua_differences.append(qua_diff) # додаємо до нового списку
    # print(f"вводимо в кв.{qua_diff}")  # виводимо квадрат різниці

# Step 4. sum - step 3 (сума квадратів різниць)

sum_qua_differences = sum(qua_differences)  # сума квадратів різниць arr

disp = sum_qua_differences / len(arr)     # дисперсія a_set

print(f"дисперсія: {disp:.2f}")             # виводимо дисперсію a_set
```

```python
qua_variance = sum_qua_differences / len(arr)
standard_deviation = qua_variance ** (1/2)
print(f"Стандартне відхилення (Середньоквадратичне відхилення) {standard_deviation:.2f}")
```

```python
variance = standard_deviation / avg_arr
print(f"Коефіцієнт варіації {variance:.2f}")
```
