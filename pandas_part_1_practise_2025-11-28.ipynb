**Завдання 1:** Оцінити середню вартість діаманта за категоріями чистоти (clarity) \
**Інструмент для вирішення:**  groupby та agg та agg function

```python
ave_price = df.groupby("clarity")['price'].agg("mean") # групуємо по категорії, ціну агрегуємо по середній

ave_price
```

**Завдання 2:** Виявити дешеві діаманти \
**Інструменти:** loc, isin, bool mask \

Дешевий це: carat > 1.2, color = "D" або "E", price < 2000

```python
mask = df['carat'] > 1.2 # зробили маску (фільтр) по каратам
df[mask]
```

```python
mask_2 = (df['price'] > 2000) # зробили маску (фільтр) по вартості
df[mask_2]
```

```python
df[(mask) & (mask_2) & ((df['color'] == "D") | (df['color'] == "E"))] # ми можемо всі фільти задати масками і потім "міксувати" маски в  залежності від задачі
# df[(df['carat'] > 1.2) & ((df['color'] == "D") | (df['color'] == "E"))]
```

**Завдання 3:** Знайди 10 найдорожчих діамантів у кожній категорії кольору.

```python
group_color = df.groupby("color")['price'].max().reset_index().sort_values("price",ascending=False).head(10)

group_color
```

```python
df_sorted = df.sort_values(by=["color","price"], ascending=[True, False])
df_sorted
```

```python
max_price = df_sorted.groupby("color").head(10)
max_price 
```
