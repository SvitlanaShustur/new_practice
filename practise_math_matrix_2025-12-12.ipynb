```python
# Task 0.1.1. У тебе є матриці розмірів: A: 3×2 B: 2×4
# Питання.Чи можна множити A · B? Який розмір результату?

A = np.array([
    [1, 2],
    [3, 4],
    [5, 6]
])

# Матриця B розміром 2×4
B = np.array([
    [1, 2, 3, 4],
    [5, 6, 7, 8]
])

# Множення A · B
C = np.dot(A, B)
print(C)
```

```python
# Task 0.1.2. Зробити множення та перевірити чи можливо його зробити?
A = np.array([
    [10, 20],
    [15, 25],
    [30, 10]
])  

B = np.array([
    [0.6],
    [0.4]
])  
print(A, B, A.shape, B.shape)
if A.shape[1] != B.shape[0]:
    raise ValueError('No equal size')
ans = A@B
print(ans, ans.shape)
# use @ Помнож матриці саме за допомогою оператора @, а не через np.dot чи цикли”

```

```python
# Task 0.3. Dot transpose matrix . Calculate diagonal
import numpy as np

# | user | sessions | watch_hours | spend |
# | ---- | -------- | ----------- | ----- |
# | 1    | 5        | 10          | 100   |
# | 2    | 10       | 20          | 200   |
# | 3    | 3        | 6           | 60    |
# | 4    | 8        | 16          | 160   |

X = np.array([
    [5, 10, 100],
    [10, 20, 200],
    [3, 6, 60],
    [8, 16, 160]
])

# use .T and @
X_T = X.T @ X
print(f'транспонована на помножена матриця {X_T}')

# use by prev task np.diag
print(f'вектор {np.diag(X_T)}')


# use np.linalg.matrix_rank 
rank = np.linalg.matrix_rank(X)
print("Ранг:", rank)


```

```python
# Calculate weights "Обчислити ваги"
weights = np.array([0.5, 0.2, 0.01]) # Створюємо масив ваг для кожної колонки X:
impact = X * weights # Множимо кожну колонку матриці X на відповідну вагу.
# Це означає: X[:,0] * 0.5, X[:,1] * 0.2, X[:,2] * 0.01
impact # Зберігаємо результат у змінну impact — це зважені ознаки
```

```python
# L2-normalize L2-нормалізація — це приведення стовпця (вектора) до довжини 1.
np.linalg.norm(X[:, 0])  # sessions # Обчислюємо L2-норму (довжину вектора) для першої колонки X — це ознака "sessions"
np.linalg.norm(X[:, 1])  # watch_hours # Обчислюємо L2-норму для другої колонки X — це ознака "watch_hours"
np.linalg.norm(X[:, 2])  # spend # Обчислюємо L2-норму для третьої колонки X — це ознака "spend"
```
