#### Логістична регресія/Logistics regression

Маємо об’єкт (рядок даних) з ознаками: x1, x2, x3, .... xn \
Модель рахує сирий скор: z = w(0) + w(1)x(1) ... w(n)(n) 

> w(0) - bias (зсув) \
> w(i) — ваги (коефіцієнти) \
> z - −∞,+∞

![](https://builtin.com/sites/www.builtin.com/files/styles/ckeditor_optimize/public/inline-images/sigmoid-activation-function-1_0.png)

````python
import pandas as pd
import numpy as np
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
````

````python
df = sns.load_dataset("iris")
df.head()
````

````python
features = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
target = "species"

df = df[features + [target]].dropna()
````

````python
X = df[features]
y = df[target]
````

````python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,
    random_state=42,
    stratify=y
)
````

````python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
````

````python
model = LogisticRegression()
model.fit(X_train_scaled, y_train)
````

````python
proba = model.predict_proba(X_test_scaled)[:, 1]
pred = model.predict(X_test_scaled)
````

````python
coef_df = pd.DataFrame({
    "feature": features,
    "coef": model.coef_[0]
})

coef_df
````

````python
coef_df["odds_multiplier"] = np.exp(coef_df["coef"])
coef_df
````

Модель має не запам’ятовувати дані, а узагальнювати закономірності


##### Underfitting (недонавчання)
Модель занадто проста і не вловлює закономірності

**Ознаки:**
> погано працює і на train, і на test \
> низьки метрики 

**Опрацювати:**
> ускладнити модель \
> додати ознаки \
> прибрати занадто сильну регуляризацію


##### Overfitting (перенавчання)
Модель запам’ятала train, але не працює на нових даних

> train: дуже добре
> test: погано

**Приклад:** Decision Tree без обмежень модель, яка вивчила напам’ять шум


**Опрацювати:**
> обмежити складність \
> більше даних \
> регуляризація \
> крос-валідація

**Варіант 1**
> тільки sex \
> accuracy = 65% і на train, і на test


**Варіант 2**
> дерево без обмежень \
> train = 99%, test = 70%


**Варіант 3**
> train = 78%, test = 76%

Linear Regression/Лінійна регресія прогнозує **ЧИСЛО**

**Приклади:**
> ціна квартири \
> час доставки \
> дохід користувача

Модель будує пряму лінію:
y = w(0) + w(1)x(1) + w(2)(x2) + ....

> w(0) - базове значення \
> w(i) — наскільки зміниться результат, якщо ознака зросте на 1

Приклад:price = 1000 * area + 5000



````python
# Імпортуємо модель лінійної регресії зі sklearn
# LinearRegression — це алгоритм для ЗАДАЧ РЕГРЕСІЇ, тобто коли ціль y є ЧИСЛОМ
# (наприклад: ціна, дохід, вага, температура, MedHouseVal тощо)
from sklearn.linear_model import LinearRegression

# Імпортуємо метрики якості для регресії
# Метрики потрібні, щоб після прогнозу оцінити, НАСКІЛЬКИ добре модель працює
from sklearn.metrics import mean_absolute_error, mean_squared_error

# mean_absolute_error (MAE) — середня абсолютна помилка:
#   MAE = середнє значення |y_true - y_pred|
# Інтуїтивно: "в середньому на скільки" модель помиляється (в одиницях цілі)

# mean_squared_error (MSE) — середня квадратична помилка:
#   MSE = середнє значення (y_true - y_pred)^2
# Інтуїтивно: сильніше штрафує ВЕЛИКІ помилки (бо різниця підноситься до квадрату)
````

````python
# df = sns.load_dataset("iris")
# df.head()
````

````python
# target — назва цільової колонки (те, що ми хочемо передбачити)
# У датасеті Iris "species" — це вид квітки: setosa / versicolor / virginica
target = "species"

# X — ознаки (features), тобто вхідні дані для моделі
# Тут ти береш 4 числові характеристики квітки:
# - sepal_length: довжина чашолистка
# - sepal_width:  ширина чашолистка
# - petal_length: довжина пелюстки
# - petal_width:  ширина пелюстки
# Подвійні дужки [[...]] важливі: так виходить DataFrame (матриця ознак), що зручно для sklearn
X = df[["sepal_length", "sepal_width", "petal_length", "petal_width"]]

# y — ціль (target), тобто правильні відповіді
# Це один стовпець Series, у якому записано вид квітки для кожного рядка
y = df[target]

````

````python
# Розбиваємо дані на навчальну (train) і тестову (test) вибірки
X_train, X_test, y_train, y_test = train_test_split(
    X, y,                 # X — ознаки (features), y — ціль (target)
    test_size=0.25,       # 25% даних підуть у тестову вибірку,
                           # 75% залишаться для навчання моделі
    random_state=42       # фіксує випадковість розбиття,
                           # щоб результат був відтворюваним
)
````

````python
# 1) Створюємо "словник відповідностей" (mapping):
# кожен клас (назва виду) перетворюємо на число
species_map = {"setosa": 0, "versicolor": 1, "virginica": 2}

# 2) Перетворюємо y_train (рядки) у числа за допомогою .map()
# Було: "setosa", "virginica"...
# Стало: 0, 2...
y_train_num = y_train.map(species_map)

# 3) Те саме робимо для тестових міток
y_test_num = y_test.map(species_map)

# 4) Створюємо модель лінійної регресії
model = LinearRegression()

# 5) Навчаємо модель:
# X_train — ознаки (4 параметри квітки)
# y_train_num — числова ціль (0/1/2)
model.fit(X_train, y_train_num)
````

````python
# model.coef_  і  model.intercept_ — це параметри, які модель "вивчила" після model.fit()

# coef_ (коефіцієнти) — показують, НА СКІЛЬКИ зміниться прогноз y,
# якщо збільшити відповідну ознаку X на 1 одиницю (за інших рівних умов).
# Тобто це "ваги" для кожної колонки в X.
#
# Якщо X має 4 колонки, то model.coef_ матиме 4 числа:
# [w1, w2, w3, w4] — для sepal_length, sepal_width, petal_length, petal_width

# intercept_ (вільний член) — це базове зміщення (bias):
# значення y, яке модель прогнозує, коли всі ознаки X дорівнюють 0.
# (часто це просто технічний параметр, бо X=0 може не мати реального сенсу)

model.coef_, model.intercept_
````

##### Ансамблі

Random Forest - багато Decision Tree

**Кожне дерево:**
> навчається на випадковій підвибірці даних \
> використовує випадкову частину ознак

**Потім:**

| Decision Tree    | Random Forest       |
| ---------------- | ------------------- |

| 1 думка          | колективне рішення  |


````python
# Імпортуємо RandomForestClassifier — модель для КЛАСИФІКАЦІЇ (коли y — класи, а не числа)
# Random Forest (випадковий ліс) = ансамбль багатьох дерев рішень (Decision Trees).
# Кожне дерево навчається на випадковій підвибірці даних і випадковому наборі ознак,
# а фінальний прогноз робиться "голосуванням" більшості дерев.
from sklearn.ensemble import RandomForestClassifier

# Імпортуємо метрику accuracy_score — "точність" класифікації.
# Accuracy = частка правильних прогнозів:
#   accuracy = (кількість правильних передбачень) / (загальна кількість передбачень)
# Наприклад, accuracy = 0.92 означає 92% правильних відповідей.
from sklearn.metrics import accuracy_score
````

````python
df.head()
````

````python
# 1) Список ознак (features), які будуть "вхідними даними" для моделі
# Це 4 числові характеристики квітки Iris:
# - sepal_length: довжина чашолистка
# - sepal_width:  ширина чашолистка
# - petal_length: довжина пелюстки
# - petal_width:  ширина пелюстки
features = ["sepal_length", "sepal_width", "petal_length", "petal_width"]

# 2) Назва цільової колонки (target) — те, що ми хочемо передбачити
# "species" — вид квітки (setosa / versicolor / virginica)
target = "species"


# 3) Залишаємо в df тільки потрібні колонки: 4 ознаки + target
# features + [target] створює список, наприклад:
# ["sepal_length", "sepal_width", "petal_length", "petal_width", "species"]
#
# dropna() видаляє рядки, де є пропуски (NaN) хоча б в одній з цих колонок
# Це важливо, бо більшість моделей sklearn не працюють з NaN "з коробки".
df = df[features + [target]].dropna()

# 4) Формуємо X — матрицю ознак (вхідні дані)
# Беремо тільки колонки з features
X = df[features]

# 5) Формуємо y — вектор цілі (правильні відповіді)
# Беремо колонку target = "species"
y = df[target]

````

````python
# Розділяємо дані на навчальну (train) і тестову (test) частини
X_train, X_test, y_train, y_test = train_test_split(
    X, y,                 # X — ознаки (features), y — ціль (класи / значення)
    test_size=0.25,       # 25% даних підуть у тест (перевірка),
                          # 75% — у train (навчання)
    random_state=42,      # "фіксуємо випадковість", щоб розбиття було однаковим
                          # при кожному запуску коду (відтворюваність)
    stratify=y            # дуже важливо для класифікації:
                          # зберігає однакові пропорції класів у train і test
                          # (наприклад, щоб setosa/versicolor/virginica були
                          # представлені приблизно в тих самих частках)
)

#Чому stratify=y корисно саме для Iris. У Iris є 3 класи. Якщо розбивати “просто випадково”, може статися так, що в тесті випадково буде менше прикладів одного класу. stratify=y цього не допускає — він робить так, щоб розподіл класів у train і test був максимально схожим. Швидка перевірка, що стратифікація спрацювала
print("Train class distribution:\n", y_train.value_counts(normalize=True))
print("Test class distribution:\n", y_test.value_counts(normalize=True))

# Після цього:
# X_train — ознаки для навчання
# X_test  — ознаки для тесту
# y_train — правильні класи для навчання
# y_test  — правильні класи для тесту
````

````python
# Створюємо модель RandomForestClassifier (випадковий ліс) для задачі КЛАСИФІКАЦІЇ
rf = RandomForestClassifier(
    n_estimators=100,   # скільки дерев буде в "лісі"
                        # 100 — популярне стартове значення:
                        # більше дерев → зазвичай стабільніше і точніше,
                        # але довше навчається

    max_depth=None,     # максимальна глибина кожного дерева
                        # None означає: дерево росте, поки це можливо
                        # (поки не закінчаться розділення або не стануть мінімальними вибірки)
                        # ⚠️ якщо даних мало або багато шуму — може бути ризик переобучення,
                        # тоді max_depth обмежують (наприклад 3, 5, 10)

    random_state=42     # фіксує випадковість:
                        # однакове "випадкове" навчання і той самий результат при повторному запуску
                        # (важливо для відтворюваності)
)

# Навчаємо модель на тренувальних даних:
# X_train — ознаки (4 виміри квітки)
# y_train — правильні класи (species: setosa/versicolor/virginica)
rf.fit(X_train, y_train)

# Після fit() модель:
# - побудує n_estimators дерев
# - кожне дерево навчиться по-своєму (на випадкових підвибірках рядків і ознак)
# - у фіналі "ліс" буде голосувати за клас при прогнозі

````

````python
# Робимо прогноз класів для тестових даних
# rf.predict() повертає передбачений клас для кожного рядка з X_test
pred = rf.predict(X_test)

# Рахуємо точність (accuracy) моделі
# accuracy_score порівнює справжні класи (y_test)
# з передбаченими класами (pred)
print("Accuracy:", accuracy_score(y_test, pred))
````

````python
# Створюємо таблицю (DataFrame) з важливістю ознак (feature importance)
importances = pd.DataFrame({
    "feature": features,                 # назви ознак (колонок), які ти подала в модель
    "importance": rf.feature_importances_  # числові оцінки важливості від RandomForest
})

# Сортуємо ознаки за важливістю: від найбільш важливої до найменш важливої
importances = importances.sort_values("importance", ascending=False)

# Виводимо таблицю
importances

````

````python
from sklearn.model_selection import GridSearchCV 
from sklearn.ensemble import RandomForestClassifier

# GridSearchCV — інструмент для підбору найкращих гіперпараметрів моделі.
# Він перебирає ВСІ комбінації параметрів із param_grid, оцінює їх через крос-валідацію (cv),
# і вибирає комбінацію, яка дає найкращий результат за заданою метрикою (scoring).

# 1) Сітка параметрів (які значення будемо пробувати)
param_grid = {
    "n_estimators": [100, 200, 250, 300],      # кількість дерев у лісі
                                                 # більше дерев → зазвичай стабільніше, але повільніше

    "max_depth": [None, 5, 10],                # максимальна глибина дерева
                                                 # None → дерево росте "до кінця"
                                                 # менша глибина → менше переобучення, але може бути недонавчання

    "min_samples_leaf": [1, 3, 5, 7, 10]       # мінімальна кількість об'єктів у ЛИСТКУ
                                                 # більше значення → дерево "грубіше", менше переобучення
}

# 2) Налаштовуємо GridSearchCV
grid = GridSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    # estimator — базова модель, яку будемо "тюнити"
    # random_state=42 — щоб результати були відтворюваними

    param_grid=param_grid,
    # param_grid — список варіантів параметрів, які треба перебрати

    cv=5,
    # cv=5 означає 5-fold cross-validation:
    # train-дані діляться на 5 частин (folds)
    # 4 частини → навчання, 1 частина → валідація
    # так повторюється 5 разів, і береться середній результат

    scoring="f1_weighted",
    # scoring — метрика, за якою обираємо найкращі параметри
    # ⚠️ ВАЖЛИВО: для багатокласової класифікації (Iris) краще вказати:
    # "f1_macro" або "f1_weighted"
    # бо просто "f1" зазвичай очікує binary-класифікацію і може викликати помилку
    # або працювати не так, як ти очікуєш

    n_jobs=-1
    # n_jobs=-1 — використовувати всі ядра процесора для паралельних обчислень
    # це сильно пришвидшує перебір
)

# 3) Запускаємо підбір параметрів
# GridSearchCV навчить багато моделей (для кожної комбінації параметрів і кожного fold)
# і збере їхні результати.
grid.fit(X_train, y_train)

# 4) Виводимо найкращі знайдені параметри
grid.best_params_
````

````python
# 1) Беремо найкращу модель, яку знайшов GridSearchCV
# best_estimator_ — це вже "готовий" RandomForestClassifier з найкращими параметрами
best_rf = grid.best_estimator_

# 2) Робимо прогноз на тестових даних (X_test)
# predict() повертає передбачений клас (species) для кожного рядка X_test
pred_best = best_rf.predict(X_test)

# 3) Рахуємо точність (accuracy) на тесті
# Порівнюємо справжні класи y_test з прогнозами pred_best
print("Accuracy:", accuracy_score(y_test, pred_best))
````
