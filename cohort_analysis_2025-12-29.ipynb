#### Когорта 
Це група користувачів, об’єднаних спільною подією у певний період часу

**Когорта** — це множина користувачів, які:
> мали однакову стартову подію \
> у чітко визначений період часу


**Приклади когорт:**
> користувачі, які зареєструвались у <month> \
> користувачі, які зробили першу покупку \
> користувачі, які пройшли онбординг \
> користувачі, які прийшли з Facebook

Основна проблема метрик, не зрозуміло в порівнянням з цим стало краще/гірше. 

**Що дає когортний аналіз:**
> де саме користувачі йдуть \
> у яких когортах стало краще \
> як змінюється поведінка з часом


**Типові випадки використання:**
> чи покращується продукт з часом? \
> які користувачі продовжують користуватись? \
> чи спрацювали зміни в продукті? \
> де основний відтік? \
> чи є проблеми з онбордингом?

| 2024-02 | 100% | 40% | 29% | 22% |

#### Часові когорти (Time-based cohorts)
Користувачі групуються за моментом стартової події:
> реєстрація \
> перша покупка \
> перша активність

Когорта = місяць реєстрації
| Cohort  | M0   | M1  | M2  |
| ------- | ---- | --- | --- |
| 2024-01 | 100% | 30% | 18% |
| 2024-02 | 100% | 38% | 25% |
---

#### Комбіновані (multi-dimensional) когорти

Когорта = (місяць реєстрації + канал + тариф)


Когортний аналіз завжди закінчується таблицею виду:
> рядки -> когорти \
> колонки -> час з моменту старту (offset) \
> значення -> метрика (users, retention, revenue)

pivot_table - зведена таблиця
```python
pd.pivot_table(
    data,
    index=...,
    columns=...,
    values=...,
    aggfunc=...
)
```
| Елемент   | У когортному аналізі          |
| --------- | ----------------------------- |
| `index`   | cohort (місяць реєстрації)    |
| `columns` | offset (місяць/тиждень життя) |
| `values`  | user_id / revenue             |
| `aggfunc` | nunique / sum                 |



````python
import pandas as pd

df = pd.DataFrame({
    "user_id": [1,1,2,3,3,4],
    "signup_month": ["2024-01","2024-01","2024-01","2024-02","2024-02","2024-02"],
    "active_month": ["2024-01","2024-02","2024-01","2024-02","2024-03","2024-02"]
})

df
````

````python
# Step 1. Підготувати данні 
df["signup_month"] = pd.to_datetime(df["signup_month"])
df["active_month"] = pd.to_datetime(df["active_month"])
````

````python
# Step 2. Порахувати offset 
df["cohort_index"] = (
    (df["active_month"].dt.year - df["signup_month"].dt.year) * 12 +
    (df["active_month"].dt.month - df["signup_month"].dt.month)
)

df
````

````python
# Step 3. Будуємо Pivot_table 
cohort_counts = pd.pivot_table(
    df,
    index="signup_month",
    columns="cohort_index",
    values="user_id",
    aggfunc="nunique"
)
cohort_counts
````

````python
# Step 4. Retention або інша метрика
cohort_sizes = cohort_counts.iloc[:, 0]

retention = cohort_counts.divide(cohort_sizes, axis=0)

retention = retention.round(3) * 100
retention
````

````python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

sns.heatmap(
    retention,
    annot=True,
    fmt=".0f",
    cmap="Blues"
)
plt.show()
````

````python
country_pivot = (
    df.groupby('Country')
      .agg(
          customers_cnt=('CustomerID', 'nunique'),      # кількість клієнтів
          sales_cnt=('InvoiceNo', 'nunique'),           # кількість продажів (інвойсів)
          revenue=('TotalPrice', 'sum')                 # виручка (додатково, дуже корисно)
      )
      .reset_index()
      .sort_values('revenue', ascending=False)          # можна сортувати інакше (див. нижче)
)

country_pivot.head(10)
````

## Загальний висновок з очищення

У процесі очищення даних було видалено значну кількість некоректних та неповних записів, записи, у яких відсутній CustomerID, тобто неможливо ідентифікувати конкретного клієнта, дублікати та операції з від’ємними кількостями або некоректними цінами. Після очищення сформовано якісний датасет, придатний для подальшого аналізу продажів і клієнтської поведінки.

````python
# Cohort Analysis (когортний аналіз)
# Мета цього блоку: для кожної покупки визначити
# 1) місяць покупки (InvoiceMonth)
# 2) місяць першої покупки клієнта (CohortMonth) — це і є "когорта" клієнта

import datetime as dt

# Функція "нормалізує" будь-яку дату до першого числа відповідного місяця.
# Наприклад: 2011-12-09 -> 2011-12-01
# Це потрібно, щоб групувати дані на рівні місяців, а не конкретних днів/годин.
def get_month(x):
    return dt.datetime(x.year, x.month, 1)

# Створюємо колонку InvoiceMonth — місяць, в якому відбулась транзакція.
# Тут ми беремо InvoiceDate і переводимо її у формат "перший день місяця".
df['InvoiceMonth'] = df['InvoiceDate'].apply(get_month)

# grouping — це групування по клієнтах.
# Для кожного CustomerID ми збираємо всі значення InvoiceMonth (місяці покупок цього клієнта).
grouping = df.groupby('CustomerID')['InvoiceMonth']

# CohortMonth — це місяць першої покупки клієнта.
# transform('min') повертає для кожного рядка той самий результат (мінімальний місяць) у межах групи клієнта.
# Тобто у всіх рядках одного клієнта буде однаковий CohortMonth — місяць, коли клієнт прийшов вперше.
df['CohortMonth'] = grouping.transform('min')
````

````python
df.head(5)
````

````python
# Функція витягує рік, місяць і день з datetime-колонки DataFrame
# Використовується для подальших розрахунків різниці між датами
# (наприклад, для визначення віку когорти або Recency в RFM)

def get_date_int(df, column):
    # Витягуємо рік з datetime-колонки (Series типу datetime64)
    year = df[column].dt.year
    
    # Витягуємо місяць з datetime-колонки
    month = df[column].dt.month
    
    # Витягуємо день місяця з datetime-колонки
    day = df[column].dt.day
    
    # Повертаємо три Series: рік, місяць і день
    return year, month, day
````

````python
# Витягуємо рік/місяць/день для місяця транзакції (InvoiceMonth)
# invoice_year, invoice_month, invoice_day — це Series, по одному значенню на кожен рядок df
invoice_year, invoice_month, invoice_day = get_date_int(df, 'InvoiceMonth')

# Витягуємо рік/місяць/день для місяця когорти (CohortMonth = місяць першої покупки клієнта)
cohort_year, cohort_month, cohort_day = get_date_int(df, 'CohortMonth')

# Різниця в роках між поточною транзакцією та місяцем першої покупки клієнта
years_diff = invoice_year - cohort_year

# Різниця в місяцях між поточною транзакцією та місяцем першої покупки клієнта
months_diff = invoice_month - cohort_month

# CohortIndex — це номер місяця відносно першої покупки клієнта (когорти)
# +1 робимо для того, щоб місяць першої покупки мав індекс 1 (а не 0)
# Приклад:
# CohortIndex = 1 -> місяць першої покупки
# CohortIndex = 2 -> наступний місяць після першої покупки
# CohortIndex = 3 -> через 2 місяці після першої покупки і т.д.
df['CohortIndex'] = years_diff * 12 + months_diff + 1

# Перевіряємо результат: дивимось перші рядки DataFrame
df.head()
````

````python
grouping = df.groupby(['CohortMonth', 'CohortIndex'])

cohort_data = grouping['CustomerID'].apply(pd.Series.nunique)

cohort_data = cohort_data.reset_index()

cohort_counts = cohort_data.pivot(index='CohortMonth',
                                 columns='CohortIndex',
                                 values='CustomerID')
````

````python
cohort_sizes = cohort_counts.iloc[:,0]
retention = cohort_counts.divide(cohort_sizes, axis=0)
retention.index=retention.index.date
````
