**Task 4.** Двоє користувачів у просторі ознак:

> u1 = (перегляди, лайки) = (10, 2) \
> u2 = (15, 5)

Потрібно:
> знайти відстань між ними (евклідова), \
> зрозуміти, хто “активніший” у загальному масштабі.

```python
# 1. Відстань між користувачами

# use np.linalg.norm
# code 
u1 = np.array([50, 10, 4])
u2 = np.array([70, 5, 1])

distance = np.linalg.norm(u1 - u2) # Обчислюємо відстань між двома векторами (користувачами)

print(f"Відстань між двома векторами {distance}")
# print(np.linalg.norm(u2))



# 2. Рівень активності кожного (довжина вектора)
# use np.linalg.norm
# code 
vector_length_u1 = np.linalg.norm(u1) # Pівень активності кожного / довжина вектора u1
vector_length_u2 = np.linalg.norm(u2) # Pівень активності кожного / довжина вектора u2

print(f"довжини векторів {vector_length_u1}, {vector_length_u2}")
```

**Task 5.** Нормалізація поведінкових фіч перед кластеризацією


```python
import numpy as np

users = np.array([
    [10, 3, 1],
    [40, 2, 0],
    [20, 5, 2],
    [15, 1, 1],
    [60, 7, 3]
])

users_norm = np.linalg.norm(users, axis=1) # Обчислюємо норму (довжину вектора) ДЛЯ КОЖНОГО користувача, axis=1 означає: обробляємо по рядках

# use linalg.norm
# users / by previos step
users_normalized = users / users_norm.reshape(-1, 1) # Нормалізація — ділимо кожен рядок на його норму, Використовуємо reshape, щоб отримати колонку (5,1). # .reshape(-1, 1) — це зміна форми масиву, щоб зробити його стовпчиком, а не рядком.
print(users_normalized)
```

```python
# або

users_norm = np.linalg.norm(users, axis=1, keepdims=True) # Обчислюємо норму (довжину вектора) ДЛЯ КОЖНОГО користувача, axis=1 означає: обробляємо по рядках. keepdims=True — це параметр у NumPy, який каже зберегти ті ж самі розміри масиву, навіть якщо ти робиш операцію, яка зазвичай їх зменшує.

# use linalg.norm
# users / by previos step
users_normalized = users / users_norm # Нормалізація — ділимо кожен рядок на його норму, Використовуємо reshape, щоб отримати колонку (5,1)
print(users_normalized)
```

**Task 5.** Оцініть наскільки користувачу підходить товар

```python
# user data 
u = (0.2, 0.8, 0.3, 0.1)

# product data
p = (0.25, 0.75, 0.2, 0.15)


import numpy as np


score = np.dot(u, p) # рахуємо скалярний добуток
print(score)

vector_length_u = np.linalg.norm(u) # рахуємо довжену вектора
vector_length_p = np.linalg.norm(p) # рахуємо довжену вектора

cos_sim = score / (vector_length_u * vector_length_p) # рахужмо косинусну схожість
print(cos_sim)
```

```python
# або 

def cos_sim(a, b):
    return np.dot(a,b) / (np.linalg.norm(a) * np.linalg.norm(b))
print(cos_sim(u,p))
```
