---

#### z-test (–¥–ª—è –ö–û–ù–í–ï–†–°–Ü–ô, 0/1)

```
p = x / n
```

---

#### –†—ñ–∑–Ω–∏—Ü—è –∫–æ–Ω–≤–µ—Ä—Å—ñ–π

```
delta = p_B - p_A
```

---

#### –û–±‚Äô—î–¥–Ω–∞–Ω–∞ (pooled) –∫–æ–Ω–≤–µ—Ä—Å—ñ—è

```
p_pooled = (x_A + x_B) / (n_A + n_B)
```

---

#### –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–∞ –ø–æ—Ö–∏–±–∫–∞ (SE)

```
SE = sqrt( p_pooled * (1 - p_pooled) * (1/n_A + 1/n_B) )
```

---

#### z-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞

```
z = delta / SE
```

---

#### p-value (–¥–≤–æ—Å—Ç–æ—Ä–æ–Ω–Ω—ñ–π —Ç–µ—Å—Ç)

```
p_value = 2 * (1 - NormalCDF( |z| ))

SE_diff = sqrt(

    p_B * (1 - p_B) / n_B

#### t-test (–¥–ª—è –ß–ò–°–õ–û–í–ò–• –ú–ï–¢–†–ò–ö)

#### –†—ñ–∑–Ω–∏—Ü—è —Å–µ—Ä–µ–¥–Ω—ñ—Ö

```

#### –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–∞ –ø–æ—Ö–∏–±–∫–∞ (Welch t-test)

```

#### t-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞

```

````python
# Task 
# –ü–æ—Ä–∞—Ö—É–≤–∞—Ç–∏ —Å–µ—Ä–µ–¥–Ω—î —É –≥—Ä—É–ø—ñ A —ñ B
# –ó–∞—Å—Ç–æ—Å—É–≤–∞—Ç–∏ Welch t-test
# –ü–æ–¥–∏–≤–∏—Ç–∏—Å—è –Ω–∞ p-value
# –ó—Ä–æ–±–∏—Ç–∏ –≤–∏—Å–Ω–æ–≤–æ–∫

import numpy as np
from scipy import stats

# –î–∞–Ω—ñ
A = np.array([10, 12, 11, 9, 10])
B = np.array([13, 14, 12, 15, 14])

# –ö—Ä–æ–∫ 1: —Å–µ—Ä–µ–¥–Ω—ñ –∑–Ω–∞—á–µ–Ω–Ω—è
mean_A = np.mean(A)# use mean
mean_B = np.mean(B)# use mean
delta = mean_B - mean_A
print("–°–µ—Ä–µ–¥–Ω—ñ–π —á–µ–∫ A:", mean_A)
print("–°–µ—Ä–µ–¥–Ω—ñ–π —á–µ–∫ B:", mean_B)
print("delta:", delta)
# –ö—Ä–æ–∫ 2: t-test
t_stat, p_value = stats.ttest_ind(A, B, equal_var=False) #–°—Ç–∞—Ç–∏—Å—Ç–∏—á–Ω–∏–π –≤–∏—Å–Ω–æ–≤–æ–∫^ –†—ñ–∑–Ω–∏—Ü—è —Å–µ—Ä–µ–¥–Ω—ñ—Ö —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–Ω–æ –∑–Ω–∞—á—É—â–∞ –°–µ—Ä–µ–¥–Ω—ñ–π —á–µ–∫ —É –≥—Ä—É–ø—ñ B —ñ—Å—Ç–æ—Ç–Ω–æ –≤–∏—â–∏–π, –Ω—ñ–∂ —É –≥—Ä—É–ø—ñ A –ô–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å –æ—Ç—Ä–∏–º–∞—Ç–∏ —Ç–∞–∫—É —Ä—ñ–∑–Ω–∏—Ü—é –≤–∏–ø–∞–¥–∫–æ–≤–æ ‚Äî ~0.2%. –Ø–∫ —Ü–µ –ø—Ä–∞–≤–∏–ª—å–Ω–æ —Å—Ñ–æ—Ä–º—É–ª—é–≤–∞—Ç–∏ —Å–ª–æ–≤–∞–º–∏ –ú–∏ –ø–æ—Ä—ñ–≤–Ω—è–ª–∏ —Å–µ—Ä–µ–¥–Ω—ñ–π —á–µ–∫ –¥–≤–æ—Ö –Ω–µ–∑–∞–ª–µ–∂–Ω–∏—Ö –≥—Ä—É–ø –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é Welch t-test. –°–µ—Ä–µ–¥–Ω—ñ–π —á–µ–∫ —É –≥—Ä—É–ø—ñ B (13.6) –∑–Ω–∞—á–Ω–æ –ø–µ—Ä–µ–≤–∏—â—É—î —Å–µ—Ä–µ–¥–Ω—ñ–π —á–µ–∫ —É –≥—Ä—É–ø—ñ A (10.4). –û—Ç—Ä–∏–º–∞–Ω–µ p-value = 0.002 —Å–≤—ñ–¥—á–∏—Ç—å –ø—Ä–æ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–Ω–æ –∑–Ω–∞—á—É—â—É —Ä—ñ–∑–Ω–∏—Ü—é.
print(t_stat)
print(p_value)

# t_stat > 0 - mean_A > mean_B
# t_stat < 0 - mean_A < mean_B
````

````python
# Task 

# 1. –ü–æ—Ä–∞—Ö—É–≤–∞—Ç–∏ –∫–æ–Ω–≤–µ—Ä—Å—ñ—é
# 2. –ó–∞—Å—Ç–æ—Å—É–≤–∞—Ç–∏ z-test –¥–ª—è –ø—Ä–æ–ø–æ—Ä—Ü—ñ–π
# 3. –ü–æ–¥–∏–≤–∏—Ç–∏—Å—è –Ω–∞ p-value
# 4. –ó—Ä–æ–±–∏—Ç–∏ –≤–∏—Å–Ω–æ–≤–æ–∫
import numpy as np
from scipy import stats
import math
from scipy.stats import norm

nA = 100
xA = 10

nB = 100
xB = 18

# –ö—Ä–æ–∫ 1. –ü–æ—Ä–∞—Ö—É–≤–∞—Ç–∏ –∫–æ–Ω–≤–µ—Ä—Å—ñ—é 
pA = xA / nA # you code 
pB = xB / nB# you code 

# –ö—Ä–æ–∫ 2: z-test
delta = pB -pA
p_pool =  (xA + xB) / (nA+nB)# you code
SE = math.sqrt(p_pool * (1 - p_pool) * (1/nA + 1/nB))
z = delta / SE # you code

p_value = 2 * (1 - stats.norm.cdf(abs(z)))# use 2 * (1 - norm.cdf(abs(z))) –Ü–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü—ñ—è p-value > 0.05: —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–Ω–æ –∑–Ω–∞—á—É—â–æ—ó —Ä—ñ–∑–Ω–∏—Ü—ñ –Ω–µ–º–∞—î, —Ç–∞–∫—É —Ä—ñ–∑–Ω–∏—Ü—é –º–æ–∂–Ω–∞ –ø–æ—è—Å–Ω–∏—Ç–∏ –≤–∏–ø–∞–¥–∫–æ–≤—ñ—Å—Ç—é
print("pA:", pA)
print("pB:", pB)
print("delta:", delta)
print("z:", z)
print("p-value:", p_value)
````

````python
# Task 1. –í–∏ —Ç–µ—Å—Ç—É—î—Ç–µ –Ω–æ–≤–∏–π —Ç–µ–∫—Å—Ç –Ω–∞ –∫–Ω–æ–ø—Ü—ñ –æ–ø–ª–∞—Ç–∏.
# –ú–µ—Ç—Ä–∏–∫–∞: purchase conversion (–∫—É–ø–∏–≤ / –Ω–µ –∫—É–ø–∏–≤) ‚Äî –±—ñ–Ω–∞—Ä–Ω–∞.

# –î–∞–Ω—ñ –∑–∞ —á–∞—Å —Ç–µ—Å—Ç—É:
# A (–∫–æ–Ω—Ç—Ä–æ–ª—å): 8 000 –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á—ñ–≤, 720 –æ–ø–ª–∞—Ç
# B (—Ç–µ—Å—Ç): 8 000 –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á—ñ–≤, 784 –æ–ø–ª–∞—Ç–∏

# –ö—Ä–æ–∫ 0. –§–æ—Ä–º—É–ª—é—î–º–æ –≥—ñ–ø–æ—Ç–µ–∑–∏ "–ö–æ–Ω–≤–µ—Ä—Å—ñ—è —É –≥—Ä—É–ø—ñ A –¥–æ—Ä—ñ–≤–Ω—é—î –∫–æ–Ω–≤–µ—Ä—Å—ñ—ó —É –≥—Ä—É–ø—ñ B.""

#  —Ä—ñ–∑–Ω–∏—Ü—ñ –Ω–µ–º–∞—î
#  —Ä—ñ–∑–Ω–∏—Ü—è —î

# –î–≤–æ—Å—Ç–æ—Ä–æ–Ω–Ω—ñ–π —Ç–µ—Å—Ç, –±–æ –∑–º—ñ–Ω–∞ –º–æ–∂–µ —è–∫ –ø–æ–∫—Ä–∞—â–∏—Ç–∏, —Ç–∞–∫ —ñ –ø–æ–≥—ñ—Ä—à–∏—Ç–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç.
````

````python
from scipy import stats
import math
from scipy.stats import norm
# –ö—Ä–æ–∫ 1. –†–∞—Ö—É—î–º–æ –∫–æ–Ω–≤–µ—Ä—Å—ñ—ó —Ç–∞ –¥–µ–ª—å—Ç–∞(pB‚Äã‚àípA)
n_A = 8000
x_A = 720

n_B = 8000
x_B = 784

p_A = x_A / n_A
p_B = x_B / n_B

delta = p_B - p_A
# –ö—Ä–æ–∫ 2. –û–±‚Äô—î–¥–Ω–∞–Ω–∞ –∫–æ–Ω–≤–µ—Ä—Å—ñ—è (p = (pa + pb) / count A + count B)
p_pool = (x_A + x_B) / (n_A + n_B)
# –ö—Ä–æ–∫ 3. –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–∞ –ø–æ—Ö–∏–±–∫–∞(SE)
SE = math.sqrt(p_pool * (1 - p_pool) * (1/n_A + 1/n_B))
# –ö—Ä–æ–∫ 4. z-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –¥–µ–ª—å—Ç–∞ / SE
z = delta / SE
# –ö—Ä–æ–∫ 5. p-value (–¥–≤–æ—Å—Ç–æ—Ä–æ–Ω–Ω—ñ–π)
p_value = 2 * (1 - stats.norm.cdf(abs(z))) # –°—Ç–∞—Ç–∏—Å—Ç–∏—á–Ω–æ –∑–Ω–∞—á—É—â–æ—ó —Ä—ñ–∑–Ω–∏—Ü—ñ –º—ñ–∂ –≥—Ä—É–ø–∞–º–∏ A —ñ B –Ω–µ –≤–∏—è–≤–ª–µ–Ω–æ. –ù–∞—è–≤–Ω–∏—Ö –¥–∞–Ω–∏—Ö –Ω–µ–¥–æ—Å—Ç–∞—Ç–Ω—å–æ, —â–æ–± —Å—Ç–≤–µ—Ä–¥–∂—É–≤–∞—Ç–∏, —â–æ –Ω–æ–≤–∏–π —Ç–µ–∫—Å—Ç –Ω–∞ –∫–Ω–æ–ø—Ü—ñ –≤–ø–ª–∏–Ω—É–≤ –Ω–∞ –∫–æ–Ω–≤–µ—Ä—Å—ñ—é. –°–ø–æ—Å—Ç–µ—Ä–µ–∂—É–≤–∞–Ω–∏–π –µ—Ñ–µ–∫—Ç –º–æ–∂–µ –±—É—Ç–∏ –∑—É–º–æ–≤–ª–µ–Ω–∏–π –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ –∫–æ–ª–∏–≤–∞–Ω–Ω—è–º–∏.
print("p_A:", p_A)
print("p_B:", p_B)
print("delta:", delta)
print("z:", z)
print("p-value:", p_value)
````

````python

import numpy as np
from scipy import stats

nA, xA = 8000, 720
nB, xB = 8000, 784

# –ö—Ä–æ–∫ 1. –†–∞—Ö—É—î–º–æ –∫–æ–Ω–Ω–≤–µ—Ä—Å—ñ—ó —Ç–∞ –¥–µ–ª—å—Ç–∞
pA = xA / nA
pB = xB / nB
# –†–æ–∑—Ä–∞—Ö—É–Ω–æ–∫ –¥—É–ª—å—Ç–∞
delta = pB - pA
print(delta)

# –ö—Ä–æ–∫ 2. –û–±'—î–¥–Ω–∞–Ω–∞ –∫–æ–Ω–≤–µ—Ä—Å—ñ—è
p_pool = (xA + xB) / (nA + nB)
print(p_pool)

# –ö—Ä–æ–∫ 3. SE -—Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–∞ –ø–æ—Ö–∏–±–∫–∞
# p_pool * (1 - p_pool)
se = np.sqrt(p_pool * (1 - p_pool) * (1/nA + 1/nB))
print(se)
#(pB-pA) / se

# –ö—Ä–æ–∫ 4. z-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –¥–µ–ª—å—Ç–∞ / SE
z = delta / se
print(z)

# –ö—Ä–æ–∫ 5. p-value (–¥–≤–æ—Å—Ç–æ—Ä–æ–Ω–Ω—ñ–π)
p_value = 2 * (1 - stats.norm.cdf(abs(z))) #cumulative distribution function
print(p_value)
````

–ú–µ—Ç—Ä–∏–∫–∞: –æ–ø–ª–∞—Ç–∞.

–î–∞–Ω—ñ:
- A: 12 000 –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á—ñ–≤, 1 020 –æ–ø–ª–∞—Ç
- B: 12 000 –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á—ñ–≤, 1 128 –æ–ø–ª–∞—Ç

–ü—Ä–æ–≤–µ–¥—ñ—Ç—å z-test (–¥–≤–æ—Å—Ç–æ—Ä–æ–Ω–Ω—ñ–π, ùõº=0.05 Œ±=0.05)
—Ç–∞ –∑—Ä–æ–±—ñ—Ç—å –≤–∏—Å–Ω–æ–≤–æ–∫. –î–æ–¥–∞—Ç–∫–æ–≤–æ –ø–æ—Ä–∞—Ö—É–π—Ç–µ CI

````python
# –ö—Ä–æ–∫ 1. –ö–æ–Ω–≤–µ—Ä—Å—ñ—ó
# –ö—Ä–æ–∫ 2. Pooled conversion
# –ö—Ä–æ–∫ 3. –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–∞ –ø–æ—Ö–∏–±–∫–∞
# –ö—Ä–æ–∫ 4. z-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞
# –ö—Ä–æ–∫ 5. p-value

````

````python
from scipy import stats
import math
from scipy.stats import norm

n_A, x_A = 12000, 1020
n_B, x_B = 12000, 1128

# –ö—Ä–æ–∫ 1. –†–∞—Ö—É—î–º–æ –∫–æ–Ω–Ω–≤–µ—Ä—Å—ñ—ó —Ç–∞ –¥–µ–ª—å—Ç–∞
p_A = x_A / n_A
p_B = x_B / n_B
# –†–æ–∑—Ä–∞—Ö—É–Ω–æ–∫ –¥—É–ª—å—Ç–∞
delta = p_B - p_A
print(delta)

# –û–±'—î–¥–Ω–∞–Ω–∞ –∫–æ–Ω–≤–µ—Ä—Å—ñ—è
p_pool = (x_A + x_B) / (n_A + n_B)
# –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–∞ –ø–æ—Ö–∏–±–∫–∞(SE)
SE = math.sqrt(p_pool * (1 - p_pool) * (1/n_A + 1/n_B))
# z-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –¥–µ–ª—å—Ç–∞ / SE
z = delta / SE
p_value = 2 * (1 - stats.norm.cdf(abs(z))) 
SE_diff = math.sqrt(
    p_A * (1 - p_A) / n_A +
    p_B * (1 - p_B) / n_B
)
# –†–∞—Ö—É—î–º–æ –°–Ü
CI_1 = (delta - 1.96 * SE_diff)*100
CI_2 = (delta + 1.96 * SE_diff)*100
# –°—Ç–∞—Ç–∏—Å—Ç–∏—á–Ω–æ –∑–Ω–∞—á—É—â–æ—ó —Ä—ñ–∑–Ω–∏—Ü—ñ –º—ñ–∂ –≥—Ä—É–ø–∞–º–∏ A —ñ B –Ω–µ –≤–∏—è–≤–ª–µ–Ω–æ. –ù–∞—è–≤–Ω–∏—Ö –¥–∞–Ω–∏—Ö –Ω–µ–¥–æ—Å—Ç–∞—Ç–Ω—å–æ, —â–æ–± —Å—Ç–≤–µ—Ä–¥–∂—É–≤–∞—Ç–∏, —â–æ –Ω–æ–≤–∏–π —Ç–µ–∫—Å—Ç –Ω–∞ –∫–Ω–æ–ø—Ü—ñ –≤–ø–ª–∏–Ω—É–≤ –Ω–∞ –∫–æ–Ω–≤–µ—Ä—Å—ñ—é. –°–ø–æ—Å—Ç–µ—Ä–µ–∂—É–≤–∞–Ω–∏–π –µ—Ñ–µ–∫—Ç –º–æ–∂–µ –±—É—Ç–∏ –∑—É–º–æ–≤–ª–µ–Ω–∏–π –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ –∫–æ–ª–∏–≤–∞–Ω–Ω—è–º–∏.
print("p_A:", p_A)
print("p_B:", p_B)
print("delta:", delta)
print("z:", z)
print("p-value:", p_value)
print("'CI_1','CI_2':", CI_1, CI_2)
````
